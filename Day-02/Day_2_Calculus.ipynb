```python
# Core imports for the whole lab
import numpy as np
import sympy as sp

x, y = sp.symbols('x y')      # symbolic variables we'll reuse
sp.init_printing()            # pretty-print symbolic math
np.random.seed(42)
print('Setup complete. SymPy', sp.__version__, '| NumPy', np.__version__)
```

```python
# -----------------------------------------------------------
# 🔹 1A. NUMERICAL DERIVATIVE (finite difference)
# -----------------------------------------------------------

# The derivative is the slope: how much f changes for a tiny step h
def f(x):
    return x**2

h = 1e-6
slope_at_3 = (f(3 + h) - f(3)) / h
print('Numerical f\'(3):', round(slope_at_3, 4), ' (exact = 6)')
```

```python
# -----------------------------------------------------------
# 🔹 1B. SYMBOLIC DERIVATIVE (SymPy)
# -----------------------------------------------------------

expr = x**2
deriv = sp.diff(expr, x)          # differentiate w.r.t. x
print('d/dx (x**2) =', deriv)     # 2*x

# Evaluate the symbolic derivative at x = 3
print('Symbolic f\'(3) =', deriv.subs(x, 3))
```

```python

def g(x):
    return x**3 + 2*x
```

```python
#Numerical derivative at x = 2 (use h = 1e-6)
h = 1e-6
slope_at_2_for_g = (g(2 + h) - g(2)) / h
print(f"Numerical g'(2): {round(slope_at_2_for_g, 4)}")
```

```python
#Symbolic derivative of x**3 + 2*x
expr_g = x**3 + 2*x
deriv_g = sp.diff(expr_g, x)          # differentiate w.r.t. x
print('d/dx (x**3 + 2*x) =', deriv_g)

# Evaluate the symbolic derivative at x = 2
print('Symbolic g\'(2) =', deriv_g.subs(x, 2))
```

### Comparison of Numerical and Symbolic Derivatives

- **Numerical derivative of `g(x)` at `x = 2`:** `14.0`
- **Symbolic derivative of `g(x)` at `x = 2`:** `14`

Both methods yield very close results, confirming the accuracy of both calculations.

```python
# -----------------------------------------------------------
# 🔹 2A. PARTIAL DERIVATIVES
# -----------------------------------------------------------

f2 = x**2 + 3*x*y + y**2

# A partial derivative differentiates ONE variable, holding others fixed
print('df/dx =', sp.diff(f2, x))     # 2*x + 3*y
print('df/dy =', sp.diff(f2, y))     # 3*x + 2*y
```

```python
# -----------------------------------------------------------
# 🔹 2B. THE GRADIENT (vector of partials)
# -----------------------------------------------------------

# The gradient stacks every partial derivative into one vector
grad = [sp.diff(f2, v) for v in (x, y)]
print('grad f =', grad)

# Evaluate the gradient at the point (x=1, y=2)
grad_at = [g.subs({x: 1, y: 2}) for g in grad]
print('grad f at (1, 2) =', grad_at)   # points in the steepest-ascent direction
```

```python
# -----------------------------------------------------------
# 🔹 3A. CHAIN RULE BY HAND vs SymPy
# -----------------------------------------------------------

# y = sin(x**2) is a composition: outer = sin(u), inner = u = x**2
# Chain rule:  dy/dx = cos(u) * du/dx = cos(x**2) * 2x
by_hand = sp.cos(x**2) * 2*x
by_sympy = sp.diff(sp.sin(x**2), x)

print('By hand :', by_hand)
print('By SymPy:', by_sympy)
print('Match?  ', sp.simplify(by_hand - by_sympy) == 0)
```

```python
# -----------------------------------------------------------
# 🔹 4A. FORWARD PASS
# -----------------------------------------------------------

# Tiny toy problem: 4 samples, 3 input features, 5 hidden units, 1 output
X = np.random.randn(4, 3)
Y = np.random.randn(4, 1)
W1 = np.random.randn(3, 5) * 0.1
W2 = np.random.randn(5, 1) * 0.1

z1 = X @ W1                 # linear layer 1
h  = np.maximum(0, z1)      # ReLU activation
y_hat = h @ W2              # linear layer 2 (prediction)
loss = ((y_hat - Y) ** 2).mean()
print('Initial loss:', round(loss, 4))
```

```python

# -----------------------------------------------------------
# 🔹 4B. BACKWARD PASS (gradients via the chain rule)
# -----------------------------------------------------------

# Work backwards from the loss, one link at a time
dy   = 2 * (y_hat - Y) / Y.size      # d loss / d y_hat
dW2  = h.T @ dy                      # d loss / d W2
dh   = dy @ W2.T                     # d loss / d h
dz1  = dh * (z1 > 0)                 # ReLU gradient (1 where z1>0 else 0)
dW1  = X.T @ dz1                     # d loss / d W1

print('dW1 shape:', dW1.shape, '(matches W1)')
print('dW2 shape:', dW2.shape, '(matches W2)')
```

```python
# -----------------------------------------------------------
# 🔹 4C. ONE GRADIENT-DESCENT STEP SHOULD LOWER THE LOSS
# -----------------------------------------------------------

lr = 0.1
W1 -= lr * dW1               # step downhill
W2 -= lr * dW2

# Recompute the loss after the update
h_new = np.maximum(0, X @ W1)
loss_new = ((h_new @ W2 - Y) ** 2).mean()
print('Loss before:', round(loss, 4))
print('Loss after :', round(loss_new, 4), '-> should be lower')
```

```python

Xb = np.random.randn(6, 4)
Yb = np.random.randn(6, 1)
Wa = np.random.randn(4, 8) * 0.1     # input -> hidden
Wb = np.random.randn(8, 1) * 0.1     # hidden -> output
```

### Forward Pass Equations:

*   **Input Layer to Hidden Layer (Linear Transformation):** `z1 = X @ W1`
*   **Activation (ReLU):** `h = np.maximum(0, z1)`
*   **Hidden Layer to Output Layer (Prediction):** `y_hat = h @ W2`
*   **Loss Calculation (Mean Squared Error):** `loss = ((y_hat - Y) ** 2).mean()`

Now, let's compute the forward pass for the new variables `Xb`, `Yb`, `Wa`, and `Wb`.

```python
# Forward pass: z1, h = ReLU(z1), y_hat, loss
z1b = Xb @ Wa
hb = np.maximum(0, z1b)
y_hat_b = hb @ Wb
loss_b = ((y_hat_b - Yb) ** 2).mean()

print('Initial loss for new network:', round(loss_b, 4))
```

### Backward Pass Equations:

*   **Loss Gradient:** `dyb = 2 * (y_hat_b - Yb) / Yb.size`
*   **Output Layer Weight Gradient:** `dWb_grad = hb.T @ dyb`
*   **Hidden Layer Activation Gradient:** `dhb = dyb @ Wb.T`
*   **Hidden Layer Linear Gradient (ReLU):** `dz1b = dhb * (z1b > 0)`
*   **Input Layer Weight Gradient:** `dWa_grad = Xb.T @ dz1b`

Let's apply these equations to `Xb`, `Yb`, `Wa`, and `Wb` to calculate the gradients.

```python
# Calculate gradients for the new network
dyb = 2 * (y_hat_b - Yb) / Yb.size
dWb_grad = hb.T @ dyb
dhb = dyb @ Wb.T
dz1b = dhb * (z1b > 0)
dWa_grad = Xb.T @ dz1b

print('dWb_grad shape:', dWb_grad.shape, '(matches Wb)')
print('dWa_grad shape:', dWa_grad.shape, '(matches Wa)')
```

```python
#One step with lr = 0.05; print loss before and after
lr_b = 0.05
Wa -= lr_b * dWa_grad
Wb -= lr_b * dWb_grad

# Recompute the loss after the update
z1b_new = Xb @ Wa
hb_new = np.maximum(0, z1b_new)
y_hat_b_new = hb_new @ Wb
loss_b_new = ((y_hat_b_new - Yb) ** 2).mean()

print('Loss before:', round(loss_b, 4))
print('Loss after :', round(loss_b_new, 4), '-> should be lower')
```

```python
import sympy as sp
x, y = sp.symbols('x y') # Re-declare x and y as independent symbolic variables

# -----------------------------------------------------------
# 🔹 5A. THE HESSIAN (matrix of second derivatives = curvature)
# -----------------------------------------------------------

f5 = x**2 + 3*x*y + y**2
H = sp.hessian(f5, (x, y))
print('Hessian of f:')
sp.pprint(H)        # [[2, 3], [3, 2]]
```

```python
# -----------------------------------------------------------
# 🔹 5B. GRADIENT DESCENT ON A SIMPLE FUNCTION
# -----------------------------------------------------------

# Minimise f(x) = (x - 4)**2, whose minimum is at x = 4.
# Update rule:  x <- x - lr * f'(x),  with f'(x) = 2*(x - 4)
xv = 0.0          # starting point
lr = 0.2
for step in range(15):
    grad = 2 * (xv - 4)        # the derivative
    xv = xv - lr * grad        # step against the gradient
print('Converged x:', round(xv, 3), ' (true minimum = 4)')
```

```python
#Hessian of x**4 + y**2
f_new = x**4 + y**2
H_new = sp.hessian(f_new, (x, y))
print('Hessian of f_new:')
sp.pprint(H_new)
```

```python
#Gradient descent to minimise (x - 7)**2
xv = 0.0
lr = 0.1
for step in range(20):
    grad = 2 * (xv - 7) # Derivative of (x - 7)**2
    xv = xv - lr * grad  # Update rule
print('Final x:', xv)
```
